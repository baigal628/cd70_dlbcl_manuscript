```r
getwd()
setwd("/liulab/galib/dlbcl_manuscript/")
library(tidyverse)
library(Seurat)
library(here)
library(harmony)
library(viridis)
library(RColorBrewer)
library(Polychrome)
PurpleAndYellow()
library(ComplexHeatmap)
library(devtools)
library(presto)
library(dplyr)
library(ggplot2)
library(ggpubr)
```

```r
B_cell<- readRDS("./data/objects/B_cell_final.rds")
```

```r
malignant_cells<- read.table('data/infercnv/malignant_cells.tsv', header = FALSE, sep = '\t')  %>% pull('V1')infercnv_obj<- readRDS("/liulab/galib/mouse_scRNAseq_margaret/data/infercnv/inferCNV_out_new/Pool105_75/run.final.infercnv_obj")
```

```r
all_poolid_no_sick<- B_cell@meta.data  %>% filter(genotype != 'WT', age != 'sick')  %>% 
    count(pool_id, genotype, age)  %>% arrange(age, genotype)

all_poolid_no_sick$genotype_new<- gsub(all_poolid_no_sick$genotype, pattern = '/[-+]', replacement = "")  %>% 
    gsub(pattern = ';', replacement = "_")  %>% gsub(pattern = '-', replacement = "")
```

```r
inpath = '/liulab/galib/mouse_scRNAseq_margaret/data/infercnv/inferCNV_out_new/'
```

```r
B_cell@meta.data  %>% filter(sample_id %in% c('191', '004', '109', '90b'))  %>% 
    count("pool_id", "sample_id", "genotype", "age")
```

```r
for (i in c(1:nrow(all_poolid_no_sick))){
    
    poolid = all_poolid_no_sick$pool_id[i]
    genotype = all_poolid_no_sick$genotype[i]
    age = all_poolid_no_sick$age[i]
    genotype_new = all_poolid_no_sick$genotype_new[i]
    
    message('Reading ', paste0(inpath, poolid, '/run.final.infercnv_obj ...'))

    infercnv_obj<- readRDS(paste0(inpath, poolid, '/run.final.infercnv_obj'))

    mtx_obs<- infercnv_obj@expr.data[,infercnv_obj@observation_grouped_cell_indices[[genotype]]]
    mtx_ref<- infercnv_obj@expr.data[,infercnv_obj@reference_grouped_cell_indices[['WT']]]
    mtx_obs = t(mtx_obs)

    message('ncells: ', dim(mtx_obs)[1], ', ngenes: ', dim(mtx_obs)[2])


    gene_position = infercnv_obj@gene_order


    row_annotation = data.frame(row.names = rownames(mtx_obs), infercnv = rep('non-malignant', nrow(mtx_obs)))
    row_annotation[rownames(row_annotation) %in% malignant_cells, 'infercnv'] = 'malignant'
    row_annotation =  row_annotation %>% arrange(infercnv)

    right_annotation <- rowAnnotation(infercnv = row_annotation$infercnv,
                                      col = list("infercnv" = c("malignant" = '#FF7F00', "non-malignant"= '#984EA3')))

    cnv_col_fun<- circlize::colorRamp2(c(1 - max(abs(mtx_obs - 1)), 1, 1 + max(abs(mtx_obs - 1))), c("blue", "white", "red"))
    cnv_col_fun<- circlize::colorRamp2(c(0.8, 1, 1.2), c("blue", "white", "red"))

    ht = Heatmap(mtx_obs[rownames(row_annotation),],
            name = poolid, 
            border_gp = gpar(col = "black"),
            column_split = factor(gene_position[colnames(mtx_obs), 'chr'], levels = unique(gene_position[['chr']])),
            column_title_rot = 90,
            column_title_gp = gpar(fontsize = 6),
            column_gap = unit(0.1, "mm"),
            cluster_rows = FALSE, cluster_columns = FALSE, 
            show_row_names = FALSE, show_column_names = FALSE, 
            right_annotation = right_annotation,
            show_heatmap_legend = TRUE,
            col = cnv_col_fun)
    

    pdf(paste0('./results/figures/10_inferCNV_heatmap_', poolid, '_', genotype_new, '_', age, '.pdf'), width = 4, height = 2.5)
    draw(ht)
    dev.off()
}
```
